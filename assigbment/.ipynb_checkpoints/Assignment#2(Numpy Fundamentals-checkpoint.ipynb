### Convert a 1D array to a 2D array with 2 rows?

#### Desired output::

## Question:2

###  How to stack two arrays vertically?

#### Desired Output::

```python
arr1=np.arange(10).reshape(2,5)
arr2=np.ones(10).reshape(2,5)
arr3=arr2.astype(np.int32)
arr3
```

```python
np.vstack((arr1,arr3))
```

## Question:3

### How to stack two arrays horizontally?

#### Desired Output::

```python
arr1=np.arange(10).reshape(2,5)
arr2=np.ones(10).reshape(2,5)
arr3=arr2.astype(np.int32)
np.hstack((arr1,arr3))
```
