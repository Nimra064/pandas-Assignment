HOW TO COMBINE BOTH ARRAY TO EACH OTHER
COLUMN WISE COMBINE MEANS HORIZENTAL STACKING
ROW WISE COMBINE MEANS VERTICAL STACKING

```python
np.vstack((arr1,arr2))  #row wise combination, if we want to combine row wise its means rows of both arrays are equal to each other
```

```python
np.hstack((arr1,arr2))  #column wise stacking,Error indicate that column_arr1!= column_arr2 its must be equal to each other for combination
```

```python
arr3=np.concatenate((arr1,arr2) , 1)#concatenate Function Basically use for joining two or more arrays but difference size
arr3
```
